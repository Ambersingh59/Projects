### Machine learning model build using sickrit libaries

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,classification_report
from sklearn.ensemble import RandomForestClassifier,BaggingClassifier,AdaBoostClassifier,GradientBoostingClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
```

```python
x=df.drop('Diagnosis',axis=1)
y=df['Diagnosis']
```

```python
models=[LogisticRegression(),RandomForestClassifier(),AdaBoostClassifier(),GradientBoostingClassifier(),SVC(),DecisionTreeClassifier()
        ,KNeighborsClassifier(),GaussianNB(),BaggingClassifier()]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=58,test_size=0.2)
```

```python
for i,models in enumerate(models):
    models.fit(x_train,y_train)
    pred=models.predict(x_test)
    print(models)
    print(accuracy_score(y_test,pred))
    print(classification_report(y_test,pred))
    print()
```

## Resampling

```python
from imblearn.over_sampling import RandomOverSampler,SMOTE
from imblearn.pipeline import Pipeline
```

```python
resample=[RandomOverSampler(),SMOTE()]
```

```python
for res in resample:
    for mod in models:
        pipe=Pipeline([('resampling',res),
              ('modeling',mod)])

        pipe.fit(x_train,y_train)
        predp=pipe.predict(x_test)
        print(mod)
        print(res)
        print(accuracy_score(y_test,predp))
        print(classification_report(y_test,predp))
        print()
```

```python
from sklearn.metrics import roc_auc_score,roc_curve
```

```python
models=[LogisticRegression(),RandomForestClassifier(),AdaBoostClassifier(),GradientBoostingClassifier(),SVC(probability=True),DecisionTreeClassifier()
        ,KNeighborsClassifier(),GaussianNB(),BaggingClassifier()]
```

```python
plt.figure(figsize=(12,5))
for i in (models):
    i.fit(x_train,y_train)
    pred=i.predict_proba(x_test)[:,-1]
    fpr,tpr,threshold=roc_curve(y_test,pred)
    auc=np.round(roc_auc_score(y_test,pred),3)
    plt.plot(fpr,tpr,label=f'{i}={auc}')
    plt.legend()
```
