## Machine learning model build using sickirt-learn libaries

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
```

```python
df.head(4)
```

```python
from sklearn.ensemble import RandomForestRegressor,BaggingRegressor,GradientBoostingRegressor,AdaBoostRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import cross_val_score,KFold,StratifiedKFold,LeaveOneOut,ShuffleSplit
```

### creating dummies

```python
del df['model']
```

```python
new_df=pd.get_dummies(df)
```

```python
new_df.head()
```

```python
x=new_df.drop(columns=['price'])
y=new_df['price']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
pred=model.predict(x_test)
print(mean_squared_error(y_test,pred))
print(r2_score(y_test,pred))
```

```python
model=DecisionTreeRegressor()
model.fit(x_train,y_train)
pred=model.predict(x_test)
print(mean_squared_error(y_test,pred))
print(r2_score(y_test,pred))
```

```python
model=RandomForestRegressor()
model.fit(x_train,y_train)
pred=model.predict(x_test)
print(mean_squared_error(y_test,pred))
print(r2_score(y_test,pred))
```

```python
model=[BaggingRegressor(),GradientBoostingRegressor(),AdaBoostRegressor(),SVR()]
```

```python
for i, model in enumerate(model):
    print(model)
    model.fit(x_train, y_train)
    pred = model.predict(x_test)
    print(mean_squared_error(y_test,pred))
    print(r2_score(y_test,pred))
    print()
```

### Cross validation

```python
scores=cross_val_score(estimator=RandomForestRegressor(),X=x,y=y,cv=KFold(10))
print(scores)
print(scores.mean())
```

```python
modell=[BaggingRegressor(),GradientBoostingRegressor(),AdaBoostRegressor()]
```

```python
for i,j  in enumerate(modell):
    print(j)
    scores=cross_val_score(estimator=j,X=x,y=y,cv=KFold(10))
    print(scores)
    print(scores.mean())
    print()
```
