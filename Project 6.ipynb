## Machine learning model build using sickirt-learn libaries

#### Observation:The data set is a classfication data,the dependent feature is 'Recurred' and remaings columns are the independent features


###### Note:In this data set some columns are objects so i creating dummies ab0ve data

```python
data=pd.get_dummies(df)
```

```python
data.head()
```

### importing libaries

```python
from sklearn.metrics import accuracy_score,classification_report
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier,AdaBoostClassifier,GradientBoostingClassifier,BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
```

```python
x=data.drop('Recurred',axis=1)
y=data['Recurred']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=47)
```

```python
models=[LogisticRegression(),RandomForestClassifier(),DecisionTreeClassifier(),AdaBoostClassifier(n_estimators=50),
        GradientBoostingClassifier(),SVC(),BaggingClassifier(),KNeighborsClassifier(n_neighbors=5),GaussianNB()]
```

```python
for i,models in enumerate(models):
    models.fit(x_train,y_train)
    pred=models.predict(x_test)
    print(models)
    print(accuracy_score(y_test,pred))
    print(classification_report(y_test,pred))
    print()
```

#### Observation:Models are perform well for this data set

```python
for i,models in enumerate(models):
    models.fit(x_train,y_train)
    pred=models.predict(x_train)
    print(models)
    print(accuracy_score(y_train,pred))
    print(classification_report(y_train,pred))
    print()
```

## Cross validation method

```python
from sklearn.model_selection import StratifiedKFold,KFold,LeaveOneOut,ShuffleSplit,cross_val_score
```

```python
for i,j in enumerate(models):
    scores=cross_val_score(estimator=j,X=x,y=y,cv=LeaveOneOut())
    print(j)
    print(scores)
    print(scores.mean())
    print()
```

### Resampling method

```python
from imblearn.pipeline import Pipeline
from imblearn.over_sampling import RandomOverSampler,SMOTE,ADASYN
from imblearn.under_sampling import RandomUnderSampler,EditedNearestNeighbours
```

```python
model=[LogisticRegression(),DecisionTreeClassifier(),
        KNeighborsClassifier()]
```

```python
resampling=[RandomOverSampler()]
```

```python
for res in resampling:
    for mod in model:
        pipe=Pipeline([('resampling',res),('modeling',mod)])
```

```python
pipe.fit(x_train,y_train)
y_pred=pipe.predict(x_test)
print(res)
print(mod)
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
print()
```

### ROC and AUC

```python
from sklearn.metrics import roc_curve,roc_auc_score
```

```python
models=[LogisticRegression(),RandomForestClassifier(),DecisionTreeClassifier(),AdaBoostClassifier(),
        GradientBoostingClassifier(),SVC(probability=True),BaggingClassifier(),KNeighborsClassifier(),GaussianNB()]
```

```python
plt.figure(figsize=(10,6))
for i in models:
    i.fit(x_train,y_train)
    y_prob=i.predict_proba(x_test)[:,1]
    fpr,tpr,threshold=roc_curve(y_test,y_prob)
    auc=np.round(roc_auc_score(y_test,y_prob),3)
    plt.plot(fpr,tpr,label=f'Auc for {i} is={auc}')
    plt.legend()
     
```

#### For this model Logisticregression is perform best
